### Step Function
$$
h(x) = \begin{cases} 
0 & (x \le 0) \\ 
1 & (x > 0) 
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return (x > 0).astype(int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### Sigmoid Function

$$
h(x) = \frac{1}{1 + \exp(-x)}
$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### ReLU (Rectified Linear Unit)

$$
h(x) = \begin{cases} 
x & (x > 0) \\ 
0 & (x \le 0) 
\end{cases}
$$

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.5, 5.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return (x > 0).astype(int)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)

plt.figure(figsize=(10, 6))
plt.plot(x, step_function(x), label='Step', linestyle='--')
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.ylim(-0.5, 5.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Activation Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

$$
\mathbf{x}=
\begin{pmatrix}
x_1 & x_2
\end{pmatrix}\in\mathbb{R}^{1\times 2},\quad
\mathbf{W}^{(1)}\in\mathbb{R}^{2\times 3},\quad
\mathbf{b}^{(1)}=
\begin{pmatrix}
b_1^{(1)} & b_2^{(1)} & b_3^{(1)}
\end{pmatrix}\in\mathbb{R}^{1\times 3},\\
\mathbf{a}^{(1)}=\mathbf{x}\mathbf{W}^{(1)}+\mathbf{b}^{(1)}\in\mathbb{R}^{1\times 3}.\\
$$
$$
\mathbf{W}^{(1)}=
\begin{pmatrix}
w_{11}^{(1)} & w_{21}^{(1)} & w_{31}^{(1)}\\
w_{12}^{(1)} & w_{22}^{(1)} & w_{32}^{(1)}
\end{pmatrix},\quad
\mathbf{a}^{(1)}=
\begin{pmatrix}
a_1^{(1)} & a_2^{(1)} & a_3^{(1)}
\end{pmatrix}.
$$
$$
a_j^{(1)}=\sum_{i=1}^{2} w_{ji}^{(1)}x_i+b_j^{(1)}.\\
a_1^{(1)}=w_{11}^{(1)}x_1+w_{12}^{(1)}x_2+b_1^{(1)}.
$$
![weights_1](../images/weights_1.png)

```python
x = np.array([1.0, 0.5])                # (2,)
W1 = np.array([[0.1, 0.3, 0.5],
               [0.2, 0.4, 0.6]])        # (2, 3)
b1 = np.array([0.1, 0.2, 0.3])          # (3,)
a1 = x @ W1 + b1
z1 = sigmoid(a1)
print(a1)
print(z1)
```

$$
\mathbf{a}^{(2)}=\mathbf{z}^{(1)}\mathbf{W}^{(2)}+\mathbf{b}^{(2)}.\\
a_k^{(2)}=\sum_{j=1}^{3} w_{kj}^{(2)}\, z_j^{(1)}+b_k^{(2)}.
$$
![weights](../images/weights_2.png)

```python
W2 = np.array([[0.1, 0.4],
               [0.2, 0.5],
               [0.3, 0.6]])             # (3, 2)
b2 = np.array([0.1, 0.2])               # (2,)
a2 = z1 @ W2 + b2
z2 = sigmoid(a2)
print(a2)
print(z2)
```

$$
\mathbf{a}^{(3)}=\mathbf{z}^{(2)}\mathbf{W}^{(3)}+\mathbf{b}^{(3)}.\\
\mathbf{y}=\sigma\!\left(\mathbf{a}^{(3)}\right)=\mathbf{a}^{(3)}
$$
![weights_3.png](../images/weights_3.png)

```python
def identity_function(x):
    return x

W3 = np.array([[0.1, 0.3],
               [0.2, 0.4]])             # (2, 2)
b3 = np.array([0.1, 0.2])               # (2,)
a3 = z2 @ W3 + b3
y = identity_function(a3)
print(a3)
print(y)
```

```python
def init_network():
    network = {}
    network["W1"] = np.array([[0.1, 0.3, 0.5],
                              [0.2, 0.4, 0.6]])     # (2, 3)
    network["b1"] = np.array([0.1, 0.2, 0.3])       # (3,)
    network["W2"] = np.array([[0.1, 0.4],
                              [0.2, 0.5],
                              [0.3, 0.6]])           # (3, 2)
    network["b2"] = np.array([0.1, 0.2])             # (2,)
    network["W3"] = np.array([[0.1, 0.3],
                              [0.2, 0.4]])           # (2, 2)
    network["b3"] = np.array([0.1, 0.2])             # (2,)
    return network

def forward(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]

    a1 = x @ W1 + b1
    z1 = sigmoid(a1)

    a2 = z1 @ W2 + b2
    z2 = sigmoid(a2)

    a3 = z2 @ W3 + b3
    y = identity_function(a3)
    return y

network = init_network()
x = np.array([1.0, 0.5])          # (2,)
y = forward(network, x)
print(y)
```

### Softmax

$$
y_k=\frac{\exp(a_k)}{\sum_{i=1}^{n}\exp(a_i)}
$$

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

a = np.array([1010, 1000, 990])
y = softmax(a)
print(y)
```

$$
y_k=\frac{\exp(a_k)}{\sum_{i=1}^{n}\exp(a_i)}
=\frac{C\,\exp(a_k)}{C\sum_{i=1}^{n}\exp(a_i)}
=\frac{\exp(a_k+\log C)}{\sum_{i=1}^{n}\exp(a_i+\log C)}
=\frac{\exp(a_k+C')}{\sum_{i=1}^{n}\exp(a_i+C')}.
$$
$$
y_k=\frac{\exp(a_k-\max(a))}{\sum_{i=1}^{n}\exp(a_i-\max(a))}.
$$

```python
def softmax(a):
    c = np.max(a, axis=-1, keepdims=True)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a, axis=-1, keepdims=True)
    y = exp_a / sum_exp_a
    return y

a = np.array([1010, 1000, 990])
y = softmax(a)
print(y)
print(np.sum(y))
```

```python
import pickle

def get_test_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=False)
    return x_test, t_test

def init_network():
    with open("../models/sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    return network

def predict(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]
    a1 = x @ W1 + b1
    z1 = sigmoid(a1)
    a2 = z1 @ W2 + b2
    z2 = sigmoid(a2)
    a3 = z2 @ W3 + b3
    y = softmax(a3)
    return y

x, t = get_test_data()
network = init_network()
y = predict(network, x)
p = np.argmax(y, axis=-1)
accuracy = np.mean(p == t)
print("Accuracy: " + str(float(accuracy)))
```

```python
x, t = get_test_data()
network = init_network()
batch_size = 100
accuracy_cnt = 0

for i in range(0, len(x), batch_size):
    x_batch = x[i:i + batch_size]
    t_batch = t[i:i + batch_size]

    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=-1)
    accuracy_cnt += np.sum(p == t_batch)

print("Accuracy: " + str(float(accuracy_cnt) / len(x)))
```
