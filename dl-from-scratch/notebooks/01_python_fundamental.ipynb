```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
print(A, np.ndim(A), A.shape, A.shape[0])
```

```python
B = np.array([[5, 6], [7, 8]])
np.dot(A, B)
```

```python
C = np.array([[10, 20]])
A * C
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
